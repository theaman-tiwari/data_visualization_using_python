```python
import pandas as pd

df1 = pd.read_excel("E:\Cointab\Company X - Order Report.xlsx")
df1
```

```python
df1.rename(columns={"ExternOrderNo":"Order ID"}, inplace=True)
```

```python
df2=pd.read_excel("E:\Cointab\Company X - Pincode Zones.xlsx")
df2
```

```python
df3=pd.read_excel("E:\Cointab\Company X - SKU Master.xlsx")
df3
```

```python
df4=pd.read_excel("E:\Cointab\Courier Company - Invoice.xlsx")
```

```python
df4
```

```python
df4["Zone"] = df4["Zone"].astype(str).str.upper()
```

```python
df5=pd.read_excel("E:\Cointab\Courier Company - Rates.xlsx")
```

```python
df5
```

```python
company_x=df1.merge(df3,how="right",on=["SKU"])
company_x
```

```python
courier_comp=df4.merge(df5,how="inner",on=["Zone"])
courier_comp
```

```python
final_result=company_x.merge(courier_comp,how="inner",on=["Order ID"])
final_result
```

```python
final_result.drop_duplicates(inplace=True)
final_result.info()
```

```python
final_result["Weight (g)"]=final_result["Weight (g)"]/1000
```

```python
final_result[["Order ID","AWB Code"]].drop_duplicates(inplace=True)
```

```python
Result = pd.read_excel(r"E:\reactProject\aman-portfolio\Final Assignment Data\New Folder\Cointab_final_result.xlsx")
```

```python
Result["Order ID"]=final_result["Order ID"]
```

```python
Result["AWB Number"]=final_result["AWB Code"]
```

```python
Result["Total weight as per X (KG)"]=final_result["Weight (g)"]*final_result["Order Qty"]
```

```python
Result["Weight slab charged by Courier Company (KG)"]=final_result["Charged Weight"]
```

```python
Result["Delivery Zone as per X"]=final_result["Zone"]
```

```python
Result["Delivery Zone charged by Courier Company"]=final_result["Zone"]
```

```python
Result["Expected Charge as per X (Rs.)"]=Result["Total weight as per X (KG)"]*final_result["Charged Weight"]*100
```

```python
Result["Charges Billed by Courier Company (Rs.) "]=final_result["Billing Amount (Rs.)"]
```

```python
Result["Difference Between Expected Charges and Billed Charges (Rs.)"]=Result["Charges Billed by Courier Company (Rs.) "]-Result["Expected Charge as per X (Rs.)"]
```

```python
Result
```

```python
sum = pd.read_excel(r"E:\Cointab\Expected_Result.xlsx")
```

```python
sum = pd.DataFrame({
    'Total orders where X has been correctly charged': [len(Result[Result['Difference Between Expected Charges and Billed Charges (Rs.)'] == 0])],
    'Total Orders where X has been overcharged': [len(Result[Result['Difference Between Expected Charges and Billed Charges (Rs.)'] > 0])],
    'Total Orders where X has been undercharged': [len(Result[Result['Difference Between Expected Charges and Billed Charges (Rs.)'] < 0])],
    'Total invoice amount': [Result['Charges Billed by Courier Company (Rs.) '].sum()],
    'Total overcharging amount': [Result[Result['Difference Between Expected Charges and Billed Charges (Rs.)'] > 0]['Difference Between Expected Charges and Billed Charges (Rs.)'].sum()],
    'Total undercharging amount': [Result[Result['Difference Between Expected Charges and Billed Charges (Rs.)'] < 0]['Difference Between Expected Charges and Billed Charges (Rs.)'].sum()]
})
```

```python
with pd.ExcelWriter('Expected_Result.xlsx') as writer:
    Result.to_excel(writer, sheet_name='Order Level Calculation', index=False)
    sum.to_excel(writer, sheet_name='Summary Table', index=False)
```
